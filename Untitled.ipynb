```python
hist_prc = hist_prc[['SYMBOL', 'SERIES', 'TIMESTAMP', 'HIGH', 'LOW',
                     'TOTTRDQTY', 'TOTTRDVAL', 'TOTALTRADES']]
```

```python
hist_prc = hist_prc.loc[hist_prc['SERIES'].isin(['EQ', 'BE', 'BL', 'BT', 'GC', 'IL'])]
```

```python
#Sort the stock prices by symbol and then Time
hist_prc["TIMESTAMP"] = pd.to_datetime(hist_prc["TIMESTAMP"])
hist_prc = hist_prc.sort_values(by=["SYMBOL", "TIMESTAMP"])
```

```python
temp = hist_prc.loc[hist_prc['SYMBOL']=='GFLLIMITED']
temp.head(100)
```

```python
hist_prc = hist_prc.reset_index()
hist_prc = hist_prc.drop('index', axis=1)
```

```python
#Calculate additional columns -> Average traded price for the day, Quantity per trade and Value per trade
hist_prc["AVG"] = hist_prc[["HIGH", "LOW"]].mean(axis=1)
hist_prc["QTY_PER_TRADE"] = hist_prc["TOTTRDQTY"]/hist_prc["TOTALTRADES"]
hist_prc["VAL_PER_TRADE"] = hist_prc["TOTTRDVAL"]/hist_prc["TOTALTRADES"]
```

```python
#Calculate first lag of all the valriables to see if the previous day details can explain the next day price movement
shifted = hist_prc[['SYMBOL', 'TIMESTAMP', 'AVG', 'TOTTRDQTY', 'TOTTRDVAL','TOTALTRADES','QTY_PER_TRADE', 
                    'VAL_PER_TRADE']].groupby("SYMBOL").shift(+1)
hist_prc = hist_prc.join(shifted.rename(columns=lambda x: x+"_LAG"))
```

```python
temp = hist_prc.loc[hist_prc['SYMBOL']=='BLUEDART']
temp.head(100)
```

```python
shifted_1 = hist_prc[['SYMBOL', 'AVG','TOTTRDQTY', 'TOTTRDVAL','TOTALTRADES','QTY_PER_TRADE', 
                      'VAL_PER_TRADE']].groupby('SYMBOL').shift(+2)
hist_prc = hist_prc.join(shifted_1.rename(columns=lambda x: x+"_LAG2"))
```

```python
temp = hist_prc.loc[hist_prc['SYMBOL']=='NETF']
temp.head(100)
```

```python
#How many days from last trade?
hist_prc["DAYS_FROM_LAST_TRADE"] = hist_prc['TIMESTAMP'] - hist_prc['TIMESTAMP_LAG']
#Previous day Movement in Total trade quantity
hist_prc["TOTTRDQTY_DIFF"] = abs((hist_prc['TOTTRDQTY_LAG']-hist_prc['TOTTRDQTY_LAG2'])/hist_prc['TOTTRDQTY_LAG2'])
#Previous day Movement in Total trade value
hist_prc["TOTTRDVAL_DIFF"] = abs((hist_prc['TOTTRDVAL_LAG']-hist_prc['TOTTRDVAL_LAG2'])/hist_prc['TOTTRDVAL_LAG2'])
#Previous day Movement in Total trade counts
hist_prc["TOTALTRADES_DIFF"] = abs((hist_prc['TOTALTRADES_LAG']-hist_prc['TOTALTRADES_LAG2'])/hist_prc['TOTALTRADES_LAG2'])
#Previous day Movement in Quantity per trade
hist_prc["QTY_PER_TRADE_DIFF"] = abs((hist_prc['QTY_PER_TRADE_LAG']-hist_prc['QTY_PER_TRADE_LAG2'])/hist_prc['QTY_PER_TRADE_LAG2'])
#previous day Movement in Value per trade
hist_prc["VAL_PER_TRADE_DIFF"] = abs((hist_prc['VAL_PER_TRADE_LAG']-hist_prc['VAL_PER_TRADE_LAG2'])/hist_prc['VAL_PER_TRADE_LAG2'])
#Previous day movement in avg traded price
hist_prc["AVG_DIFF_prev"] = abs((hist_prc['AVG_LAG']-hist_prc['AVG_LAG2'])/hist_prc['AVG_LAG2'])
```

```python
#Current day movement in average traded price
hist_prc["AVG_DIFF"] = abs((hist_prc['AVG']-hist_prc['AVG_LAG'])/hist_prc['AVG_LAG'])
```

```python
hist_prc_chart = hist_prc.loc[(hist_prc['DAYS_FROM_LAST_TRADE'].dt.days<10) & ~(hist_prc['SERIES'].isin(['BL', 'IL', 'BE']))]
hist_prc_chart = hist_prc_chart.loc[hist_prc_chart['DAYS_FROM_LAST_TRADE'].dt.days<5]
hist_prc_chart = hist_prc_chart.loc[hist_prc_chart['QTY_PER_TRADE_DIFF']<1]
hist_prc_chart = hist_prc_chart.loc[hist_prc_chart['TOTTRDQTY_DIFF']<1]
# hist_prc_chart = hist_prc_chart.loc[hist_prc_chart['SYMBOL']=='CIPLA']
```
